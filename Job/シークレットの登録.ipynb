```python
from databricks.sdk import WorkspaceClient
w = WorkspaceClient()
w.secrets.create_scope(scope="hoge-scope")
```

```python
from databricks.sdk import WorkspaceClient
w = WorkspaceClient()
w.secrets.put_secret(scope="hoge-scope", key="username", string_value="your_username")
w.secrets.put_secret(scope="hoge-scope", key="password", string_value="your_password")
```

```python
from databricks.sdk import WorkspaceClient
from databricks.sdk.service.workspace import AclPermission

w = WorkspaceClient()

# グループに READ 権限を付与
w.secrets.put_acl(
    scope="hoge-scope",
    principal="your-group-name",  # グループ名を指定
    permission=AclPermission.READ
)
```

```python
# すべての ACL を表示
w.secrets.list_acls(scope="hoge-scope")

# 特定のプリンシパルの ACL を取得
w.secrets.get_acl(scope="hoge-scope", principal="your-group-name")
```
